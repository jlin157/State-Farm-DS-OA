## Step 1 - Clean and prepare your data: 
### (Since the exact name of columns is unknown, therefore, EDA can't be performed efficiently, so I'll just skip the EDA.) 

```python
#Covert string numerical features to int numerical features
# x7 & x19
for i in range(len(df)):
    df.loc[i,'x7']=float(df.loc[i,'x7'][:-1])/100 #we want to get rid of the % sign.
    df.loc[i,'x19']=round(float(df.loc[i,'x19'][1:]),2) #we want to get rid of the $ sign.
```

```python
df[['x7','x19']]
```

```python
#Missing Value
missing_count = df.isnull().sum().sort_values(ascending=False)
missing_percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
df_missing = pd.concat([missing_count, missing_percent], axis=1, keys=['Missing_Count', 'Missing_Percent'])
df_missing=df_missing[df_missing['Missing_Count']!=0]
print(len(df_missing))
#Out of the 100 features there are 42 feature with missing value
```

```python
df_missing
```

```python
df_missing.index.sort_values()
```

### Since we have 100 features and we don't know exact what are they, therefore we can be very strict on selecting useful feature. First of all, I'm going to drop columns that have more than 20% missing value. 

```python
df=df.drop(columns=df_missing[df_missing['Missing_Percent']>=0.2].index)
```

### x24 and x33 (gender & state) contain thousands of missing values in both train dataset and test dataset. With the requirement that we need to output a prediction for each of the rows in the test dataset, we can't just drop rows with missing gender or state value. 
### I personally think gender and state are values that it's missing value can't be impute by majority value nor treated as another category. Therefore I'll just drop x24 (gender) and x33(state), since we have plenty features to use.
### Similar reason applies to x79 as well, I'm uncertain about whether x79 is a categorical value or a numerical value with just two distinct numbers (0 or 1). I won't impute anything if I'm unsure what is x79 represents.

```python
df=df.drop(columns=['x24','x33','x79'])
```

### After droping x24, x33, x79, all other columns should be numerical features. Just try to be conservative, I'll use mean imputation for any missing numerical data. 

```python
df.fillna(df.mean(), inplace = True)
```

```python
#Check if I missed any missing data
df.isnull().sum().sort_values(ascending=False)
```

```python
#Encoding categorical features
#x59,x79 and x98 are possible categorical features as well, but they are already encoded, no need to encode them again.
cat_fea=['x3', 'x31', 'x39', 'x60', 'x65', 'x93']
df = pd.get_dummies(df, columns = cat_fea, drop_first = True)
```

```python
num_fea=[]
for i in df.columns:
    if i not in cat_fea and i not in ['y','x59','x79','x98']:
        num_fea.append(i)
```

```python
X = df.drop(['y'],axis=1)
y = df[['y']]
```

## Step 2 - Build your models: 

```python
#Accuracy with KFold to check for overfitting
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
#Logistic regression
algorithms = ['lbfgs', 'liblinear']

for i in algorithms:
    acc_score = cross_val_score(LogisticRegression(solver= i,random_state= 42), X, y, cv= kf, scoring="accuracy")
    auc_score = cross_val_score(LogisticRegression(solver= i,random_state= 42), X, y, cv= kf, scoring="roc_auc")
    print(f'Average Accuracy scores {i} are: {acc_score}')
    print(f'Average Accuracy score {i}: {"{:.4f}".format(acc_score.mean())}')
    print(f'Average Accuracy scores {i} are: {auc_score}')
    print(f'Average AUC {i}: {"{:.5f}".format(auc_score.mean())}')
```

### Since we didn't try to get rid of outliers and the dataset has so many features, therefore, I think random forest model is a very useful model here. Because first of all, tree-based models are generally less affected by outliers than linear models. Secondly, random forest allow a reduction in overfitting which is caused by having too many features.

```python
#Random Forest
acc_score = cross_val_score(RandomForestClassifier(random_state= 42), X, y, cv= kf, scoring="accuracy")
auc_score = cross_val_score(RandomForestClassifier(random_state= 42), X, y, cv= kf, scoring="roc_auc")
print(f'Average Accuracy scores are: {acc_score}')
print(f'Average Accuracy score: {"{:.4f}".format(acc_score.mean())}')
print(f'Average Accuracy scores are: {auc_score}')
print(f'Average AUC: {"{:.5f}".format(auc_score.mean())}')
```
